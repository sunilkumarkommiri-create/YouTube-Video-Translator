```python
import yt_dlp
import whisper
from deep_translator import GoogleTranslator
from gtts import gTTS
import os

# Put your YouTube link here
url = "https://youtu.be/NHk7scrb_9I?si=FwznUQVoDf8526BO"

# Download audio
ydl_opts = {
    'format': 'bestaudio',
    'outtmpl': 'audio.%(ext)s',
    'noplaylist': True,
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])

# Detect downloaded file
audio_file = None
for file in os.listdir():
    if file.startswith("audio"):
        audio_file = file

print("Audio file:", audio_file)

# Convert speech to text
model = whisper.load_model("tiny")  # faster
result = model.transcribe(audio_file)

text = result["text"]
print("\nOriginal:\n", text)

# Translate
translated_text = GoogleTranslator(source='auto', target='te').translate(text)
print("\nTranslated:\n", translated_text)

# Convert to speech
tts = gTTS(translated_text, lang='te')
tts.save("translated_audio.mp3")

# Download
from google.colab import files
files.download("translated_audio.mp3")
```
